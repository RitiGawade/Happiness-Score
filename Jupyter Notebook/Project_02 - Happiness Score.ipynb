```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset
df_15 = pd.read_csv("2015.csv")
df_16 = pd.read_csv("2016.csv")
df_17 = pd.read_csv("2017.csv")
df_18 = pd.read_csv("2018.csv")
df_19 = pd.read_csv("2019.csv")
df_20 = pd.read_csv("2020.csv")
df_21 = pd.read_csv("world-happiness-report-2021.csv")
df_2005to2020 = pd.read_csv("world-happiness-report.csv")
```

```python
#Reading the data from year 2005 to 2020
df_2005to2020
```

```python
#Renaming the columns of 2005 to 2020 data
df_2005to2020.rename(columns = {'year':'Year',
                       'Country name': 'Country',
                        'Life Ladder':'Happiness score'}, inplace = True)
df_2005to2020
```

```python
#Dropping columns from 2005 to 2020 data
df_2005to2020 = df_2005to2020.drop(['Positive affect', 'Negative affect'], axis=1)
df_2005to2020
```

```python
#Reading the data of year 2021
df_21
```

```python
#Renaming the columns of 2021 data
df_21.rename(columns = {'Logged GDP per capita':'Log GDP per capita',
                       'Country name': 'Country', 'Regional indicator':'Region',
                        'Ladder score':'Happiness score'}, inplace = True)
df_21
```

```python
#Dropping columns from 2021 data
df_21 = df_21.drop(['Standard error of ladder score', 'upperwhisker','lowerwhisker','Ladder score in Dystopia','Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption','Dystopia + residual'], axis=1)
df_21
```

```python
#Concata both the datasets together
df_concat = pd.concat([df_2005to2020, df_21],ignore_index = True,sort = False)
print(df_concat)
```

```python
#Comparing happiness score of 2021 with previous years
sns.kdeplot(df_21['Happiness score'], shade = True, label = '2021 Happiness Score', color = 'r')
sns.kdeplot(df_2005to2020['Happiness score'], shade = True, label = '2021 Happiness Score', color = 'b')
```

```python
#Finding first 10 countries with highest mean happiness score
plt.figure(figsize = (12,6))
plt.ylabel('Happiness score')
plt.title('First 10 countries with highest mean happiness score')
df_21.groupby('Country')['Happiness score'].mean().sort_values(ascending = False).head(10).plot.bar(color = 'violet')
```

```python
#Finding first 10 countries with highest mean happiness score
plt.figure(figsize = (12,6))
plt.ylabel('Happiness score')
plt.title('Last 10 countries with lowest mean happiness score')
df_21.groupby('Country')['Happiness score'].mean().sort_values(ascending = True).head(10).plot.bar(color = 'violet')
```

```python
#Highly correlated variables pointing towards Happiness score
plt.figure(figsize=(18, 6))
heatmap = sns.heatmap(df_21.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=5);
```
